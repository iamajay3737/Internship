# 2) Write a python program to display IMDB’s Top rated 50 movies’ data (i.e. name, rating, year of release) 
and make data frame

```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get('https://www.imdb.com/list/ls055386972/?sort=user_rating,desc&st_dt=&mode=grid&page=1&ref_=ttls_vw_grd')
soup = BeautifulSoup(page.content)
name=soup.find_all('h3',class_="lister-item-header")
mname=[]
rating=[]
year=[]

for i in name:
    for j in i.find_all('a'):
        mname.append(j.text)

for j in soup.find_all('div',class_="ipl-rating-star small"):
    rating.append(j.text.replace('\n',''))

for k in soup.find_all('span',class_="lister-item-year text-muted unbold"):
    year.append(k.text.replace('(','').replace(')',''))

import pandas as pd
import numpy as np
df=pd.DataFrame({'Name':mname,'Rating':rating,'Year of Release':year})
df.index = np.arange(1,len(df)+1)
print("IMDB’s Top rated 50 movie's data")
df
```

# 3) Write a python program to display IMDB’s Top rated 50 Indian movies’ data (i.e. name, rating, year of 
release) and make data frame.

```python
from bs4 import BeautifulSoup
import requests
```

```python
indmv=requests.get('https://www.imdb.com/india/top-rated-indian-movies/?sort=ir,desc&mode=simple&page=1')
soup=BeautifulSoup(indmv.content)
name=soup.find_all('td',class_="titleColumn")
moviesn=[]
rating=[]
year=[]

for i in name:
    for j in i.find_all('a'):
        moviesn.append(j.text)

for j in soup.find_all('td',class_="ratingColumn imdbRating"):
    rating.append(j.text.replace('\n',''))

for k in soup.find_all('span',class_="secondaryInfo"):
    year.append(k.text.replace('(','').replace(')',''))

import pandas as pd
data=pd.DataFrame({'Name':moviesn,'Rating':rating,'Year of Release':year})
imdb_movies=data.set_index('Name')
print("IMDB’s Top rated 50 Indian movie's data")
imdb_movies.head(50)
```

# 4) Write s python program to display list of respected former presidents of India(i.e. Name , Term ofoffice) 
from https://presidentofindia.nic.in/former-presidents.htm and make data frame.


```python
page = requests.get('https://presidentofindia.nic.in/former-presidents.htm')
soup = BeautifulSoup(page.content)
idk = soup.find_all('div',class_="presidentListing")
aame = []
lame = []

for i in idk:
    for j in i.find_all('h3'):
        aame.append(j.text.replace('\n',''))
aame

for k in soup.find_all('div',class_="presidentListing"):
    for h in k.find_all('p'):
        lame.append(h.text)
del lame[1:9:2]

df= pd.DataFrame({'Respected President Name':aame, 'Term of office':lame})
df1=df.set_index('Respected President Name')
df1
```

# a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.

```python
page=requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
soup=BeautifulSoup(page.content)
team=[]
mtch1=[]
mtch2=[]
pnt1=[]
pnt2=[]
rating1=[]
rating2=[]

for i in soup.find_all('span',class_="u-hide-phablet"):
    team.append(i.text)

for j in soup.find_all('td',class_="rankings-block__banner--matches"):
    mtch1.append(j.text)

for k in soup.find_all('td',class_="table-body__cell u-center-text"):
    mtch2.append(k.text)
del mtch2[1:39:2]
mtch=mtch1+mtch2

for l in soup.find_all('td',class_="rankings-block__banner--points"):
    pnt1.append(l.text)

for m in soup.find_all('td',class_="table-body__cell u-center-text"):
    pnt2.append(m.text)
del pnt2[0:39:2]
pnt=pnt1+pnt2

for n in soup.find_all('td',class_="rankings-block__banner--rating u-text-right"):
    rating1.append(n.text.replace('\n',''))

for o in soup.find_all('td',class_="table-body__cell u-text-right rating"):
    rating2.append(o.text)
rating=rating1+rating2
import pandas as pd
import numpy as np
df=pd.DataFrame({'Team':team,'Matches':mtch,'Points':pnt,'Rating':rating})
print('Top 10 ODI teams in men’s cricket :')
df.index = np.arange(1,len(df)+1)
df.head(10)
```

# Q7. Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world :

i) Headline ii) Time iii) News Link

```python
page = requests.get('https://www.cnbc.com/world/?region=world')
soup=BeautifulSoup(page.content)
head = []
time = []
link = []

for i in soup.find_all('a',class_="LatestNews-headline"):
    head.append(i.text)
    
for j in soup.find_all('time',class_="LatestNews-timestamp"):
    time.append(j.text)

for k in soup.find_all('a',class_="LatestNews-headline"):
    link.append(k['href'])

import pandas as pd
df = pd.DataFrame({'Headline':head, 'Time':time, 'News Link': link})
df1=df.set_index('Headline')
df1
```

# 9) Write a python program to scrape mentioned details from dineout.co.in and make data frame

i) Restaurant name
ii) Cuisine
iii) Location
iv) Ratings
v) Image UR

```python
page = requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
soup = BeautifulSoup(page.content)
rest = []
cui = []
loc = []
rating = []
ima = []

for i in soup.find_all('div',class_="restnt-info cursor"):
    for j in i.find_all('a',class_="restnt-name ellipsis"):
        rest.append(j.text)
rest

for k in soup.find_all('div',class_="restnt-loc ellipsis"):
    loc.append(k.text)
loc

for l in soup.find_all('div',class_="detail-info"):
    cui.append(l.text)
cui

for m in soup.find_all('div',class_="restnt-rating rating-4"):
    rating.append(m.text)
rating

for n in soup.find_all('img',class_="no-img"):
    ima.append(n['data-src'])
ima

df =pd.DataFrame({'Restaurant Name':rest, 'Cuisine':cui, 'Location':loc, 'Ratings':rating, 'Image URL':ima})
df1=df.set_index('Restaurant Name')
df1
```
